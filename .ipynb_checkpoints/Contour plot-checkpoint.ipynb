Purpose of this notebook is to analyze the results obtained when changing opponents possible max scores

```python
## library imports

import pandas as pd ## for data manipulation
import numpy as np ## also for data manipulation
import matplotlib.pyplot as plt ## for plotting
from mpl_toolkits import mplot3d ## for 3D plotting 
```

```python
## reading in results 

df = pd.read_csv(r"scoredbasedloopedopponentscores.csv") ## results of looping opponent's possible maximum score
df
```

```python
## sorting values for some reason 

df = df.sort_values(by=['Score','MaxOpponentScore'],ascending=[True,True]).reset_index(drop=True)
df
```

```python
## showing matrix of win pcts based on score of the player and their opponent's maximum score value

df_pivoted = df.pivot(index="MaxOpponentScore",columns="Score",values="WinPercentage")
df_pivoted
```

```python
## contour plot

fig,ax = plt.subplots(1,figsize=(10,10))
contour = ax.contour(df_pivoted.columns,df_pivoted.index,df_pivoted,levels=10)
ax.clabel(contour, inline=1, fontsize=10)
ax.set_xlabel("Player Chosen Score",fontweight='bold')
ax.set_ylabel("Maximum Opponent Chosen Score",fontweight='bold')
plt.show()
```

```python
## creating 2d matrices of combinations of score chosen by the player and max opponent score 
xx,yy = np.meshgrid(df_pivoted.columns,df_pivoted.index)
```

```python
## 3D scatter plot of results

fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.scatter3D(df.Score,df.MaxOpponentScore,df.WinPercentage)
ax.set_xlabel('Player Threshold Score',fontweight='bold')
ax.set_ylabel('Opponent Threshold Score',fontweight='bold')
ax.set_zlabel('Player Win Percentage',fontweight='bold')
plt.show()
```
